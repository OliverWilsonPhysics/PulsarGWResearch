```python
import pandas as pd
data=pd.read_csv("/raid5/homes/owilson/Research_Project/Updated Glitch Data with dF")
data
```

```python
def Cleaner(dataframe, column):
    df=dataframe
    newcol=[]
    badindices=[]
    for i in range(len(df[column])):
        try:
            float(df[column][i])
        except ValueError:
            badindices.append(i)
    df.drop(badindices, inplace=True)
    df=df.reset_index()
    df.drop(["index"], axis = 1, inplace = True)
    for i in range(len(df[column])):
        x=float(df[column][i])
        newcol.append(x)
    df[column]=newcol
    return(df)
data=Cleaner(data,"Distance (kpc)")
```

```python
namebefore=0
massbefore=0
masses=[]
import numpy as np
for name in data.Name:
    if name==namebefore:
        mass=massbefore
        masses.append(mass[0])
    else:
        mass=(np.random.normal(1.4,0.15,1))
        masses.append(mass[0])
    massbefore=mass
    namebefore=name
data["Random Mass (Solar Masses)"]=masses
data
```
