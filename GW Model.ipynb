```python
import pandas as pd
data=pd.read_csv("/raid5/homes/owilson/Research_Project/Updated Glitch Data with dF")
data
```

```python
def Cleaner(dataframe, column):
    df=dataframe
    newcol=[]
    badindices=[]
    for i in range(len(df[column])):
        try:
            float(df[column][i])
        except ValueError:
            badindices.append(i)
    df.drop(badindices, inplace=True)
    df=df.reset_index()
    df.drop(["index"], axis = 1, inplace = True)
    for i in range(len(df[column])):
        x=float(df[column][i])
        newcol.append(x)
    df[column]=newcol
    return(df)
data=Cleaner(data,"Distance (kpc)")
```

```python
import numpy as np
masses=np.random.normal(1.4, 0.15, len(data.Name))
data["Random Mass (Solar Masses)"]=masses
data
```

```python
data["orange GW Freq. (kHz)"]=-np.sqrt((data["Random Mass (Solar Masses)"]-2.2)/(-2.83))+1.97
data["purple GW Freq. (kHz)"]=-np.sqrt((data["Random Mass (Solar Masses)"]-2.2)/(-9.95))+2.56
data["orange GW tau (s)"]=10**((0.084**(data["Random Mass (Solar Masses)"]-0.54))-0.96)
data["purple GW tau (s)"]=10**((0.044**(data["Random Mass (Solar Masses)"]-0.12))-0.985)
data["orange h0"]=7.21*10**-24*(data["F0 (Hz)"]/10)**0.5*(data["dF0 (Hz)"]/10**(-7))**0.5*(1/data["orange GW Freq. (kHz)"])*(0.1/data["orange GW tau (s)"])**0.5*(1/data["Distance (kpc)"])
data["purple h0"]=7.21*10**-24*(data["F0 (Hz)"]/10)**0.5*(data["dF0 (Hz)"]/10**(-7))**0.5*(1/data["purple GW Freq. (kHz)"])*(0.1/data["purple GW tau (s)"])**0.5*(1/data["Distance (kpc)"])
data["orange strain (Hz^0.5)"]=data["orange h0"]*np.sqrt(data["orange GW tau (s)"])
data["purple strain (Hz^0.5)"]=data["purple h0"]*np.sqrt(data["purple GW tau (s)"])
data["Glitch Energy (ergs)"]=3.95*10**40*(data["F0 (Hz)"]/10)*(data["dF0 (Hz)"]/10**(-7))
```

```python
badindices=np.where(pd.isna(data["orange h0"])==True)
data=data.drop(badindices[0])
data=data.reset_index()
data.drop(["index"], axis = 1, inplace = True)
data
```
